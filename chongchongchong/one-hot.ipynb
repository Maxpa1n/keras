```python
import numpy as np
import string
```

```python
samples = ['the cat sat on the mat.', 'ten dog ate my hoework']
```

```python
token_index = {}
```

```python
for sample in samples:
    for word in sample.split():
        print(word)
        if word not in token_index:
            token_index[word] = len(token_index) + 1
```

```python
print(token_index)
print(len(samples))
```

```python
max_length = 10
```

```python
results =np.zeros(shape=(len(samples),
                          max_length,
                          max(token_index.values())+1))
```

```python
print(results.shape)
print(results)
```

```python
for i,sample in enumerate(samples):
    for j, word in list(enumerate(sample.split()))[:max_length]:
        index = token_index.get(word)
        results[i, j, index] = 1
```

```python
print(results)
```
